```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import re

# Inisialisasi WebDriver Chrome
options = webdriver.ChromeOptions()
options.add_argument("--headless")  # Jalankan tanpa membuka browser
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

driver = webdriver.Chrome(options=options)

# Ganti dengan URL Google Maps ulasan RSUP Prof. dr. I.G.N.G. Ngoerah
url = "https://www.google.com/maps/place/Rumah+Sakit+Umum+Pusat+Prof.+dr.+I.G.N.G.+Ngoerah/@-8.676019,115.2106687,17z/data=!4m8!3m7!1s0x2dd240e1a28d740d:0x842c9d89a4a920dd!8m2!3d-8.676019!4d115.213249!9m1!1b1!16s%2Fg%2F122r2r4n?entry=ttu&g_ep=EgoyMDI1MDMxNi4wIKXMDSoASAFQAw%3D%3D"  # Ganti dengan URL Google Maps yang benar
driver.get(url)

# Tunggu halaman sepenuhnya dimuat
time.sleep(5)

# List untuk menyimpan data
reviews_data = []

try:
    # Temukan elemen scroll ulasan
    scrollable_div = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "m6QErb"))
    )

    # Looping untuk scroll dan ambil data
    total_reviews = 0
    while total_reviews < 3000:
        # Scroll ke bawah untuk memuat lebih banyak ulasan
        driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", scrollable_div)
        time.sleep(2)

        # Ambil semua ulasan yang sudah termuat
        reviews = driver.find_elements(By.CLASS_NAME, "jftiEf")

        for review in reviews:
            try:
                author = review.find_element(By.CLASS_NAME, "d4r55").text
                rating = review.find_element(By.CLASS_NAME, "kvMYJc").get_attribute("aria-label")
                text = review.find_element(By.CLASS_NAME, "wiI7pd").text
                reviews_data.append([author, rating, text])
            except:
                continue

        total_reviews = len(reviews_data)
        print(f"Total ulasan terkumpul: {total_reviews}")

        # Jika sudah cukup, hentikan
        if total_reviews >= 3000:
            break

except Exception as e:
    print(f"Terjadi kesalahan: {e}")

# Tutup browser setelah scraping selesai
driver.quit()

# Simpan ke CSV
df = pd.DataFrame(reviews_data, columns=["Nama", "Rating", "Ulasan"])
df.to_csv("dataset/final-dataset.csv", index=False, encoding="utf-8")


print("✅ Data berhasil disimpan dalam google_reviews_rs_ngoe.csv")
```
